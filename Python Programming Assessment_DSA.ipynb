Q-1. slinding window anomaly score

```python
import heapq
from collections import defaultdict
```

```python
import heapq
from collections import defaultdict

def sliding_window_anomaly(arr, k):
    low = []    
    high = []   
    delayed = defaultdict(int)
    low_size = high_size = 0

    def prune(heap):
        while heap:
            num = -heap[0] if heap is low else heap[0]
            if delayed[num] > 0:
                delayed[num] -= 1
                heapq.heappop(heap)
            else:
                break

    def balance():
        nonlocal low_size, high_size
        if low_size > high_size + 1:
            heapq.heappush(high, -heapq.heappop(low))
            low_size -= 1
            high_size += 1
            prune(low)
        elif low_size < high_size:
            heapq.heappush(low, -heapq.heappop(high))
            high_size -= 1
            low_size += 1
            prune(high)

    def median():
        return -low[0] if k % 2 else (-low[0] + high[0]) / 2

    
    for i in range(k):
        heapq.heappush(low, -arr[i])
        low_size += 1
        heapq.heappush(high, -heapq.heappop(low))
        low_size -= 1
        high_size += 1
        balance()

    res = []

    for i in range(k, len(arr)):
        res.append(abs(arr[i] - median()))

        out_num = arr[i - k]
        in_num = arr[i]

        delayed[out_num] += 1

        if out_num <= median():
            low_size -= 1
            if out_num == -low[0]:
                prune(low)
        else:
            high_size -= 1
            if high and out_num == high[0]:
                prune(high)

        if in_num <= median():
            heapq.heappush(low, -in_num)
            low_size += 1
        else:
            heapq.heappush(high, in_num)
            high_size += 1

        balance()

    return res
```

```python
arr = [10, 20, 30, 40, 100]
k = 3
print(sliding_window_anomaly(arr, k))
```

question-2: Top-K correlated feature pairs

```python
import math

def top_k_correlated_features(data, k):
    n = len(data)
    m = len(data[0])

    features = list(zip(*data))

    means = []
    stds = []

    for f in features:
        mean = sum(f) / n
        variance = sum((x - mean) ** 2 for x in f) / n
        std = math.sqrt(variance)
        means.append(mean)
        stds.append(std)

    correlations = []

    for i in range(m):
        for j in range(i + 1, m):
            cov = sum(
                (features[i][t] - means[i]) * (features[j][t] - means[j])
                for t in range(n)
            ) / n

            if stds[i] == 0 or stds[j] == 0:
                corr = 0
            else:
                corr = cov / (stds[i] * stds[j])

            correlations.append((i, j, round(abs(corr), 3)))

    correlations.sort(key=lambda x: x[2], reverse=True)
    return correlations[:k]
```

```python
data = [
    [1, 2, 3],
    [2, 2, 6],
    [3, 6, 9],
    [4, 8, 12]
]

print(top_k_correlated_features(data, 1))
```
