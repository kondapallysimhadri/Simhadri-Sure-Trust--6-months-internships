2 Factorial using loop

```python
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
print(fact)
```

3 Function to check prime

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            return False
    return True
print(is_prime(17))
```
