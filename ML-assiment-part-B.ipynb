# Question 3: Hypothesis Testing + ML Decision Making

1) Null and Alternate Hypotheses

Null Hypothesis (H₀):
The new teaching method does not improve student performance.
(Mean of Group A = Mean of Group B)

Alternate Hypothesis (H₁):
The new teaching method improves student performance.
(Mean of Group B > Mean of Group A)

2) Z-test or T-test? Why?

We choose a Z-test

Because:

Sample size is large (n = 100 for both groups)

Data is approximately normally distributed

Population standard deviations are known

3) Decision process at 5% significance level

Significance level (α) = 0.05

Compare p-value with α

If p ≤ 0.05 → Reject H₀

If p > 0.05 → Fail to reject H₀

4) If p-value = 0.03, what is the conclusion?

Since 0.03 < 0.05

We reject the Null Hypothesis

Conclusion:
The new teaching method shows a statistically significant improvement

5) Why ML model accuracy decreases even after improvement & how to fix it

Why it happens:

New data distribution may be different

Model may overfit old data

Feature patterns changed

Noise increased in new dataset

How to fix it:

Retrain model with updated data

Perform feature re-engineering

Apply regularization

Use cross-validation

Monitor data drift continuously

# Question 4: Model Selection Under Business Constraints

1) Which model would you choose and why?

Logistic Regression / Linear SVM

Reasons:

Fast prediction (real-time)

Works well on large datasets

Interpretable (important in banking)

Can handle imbalance with class weights

2) Most important metric and why

Recall is the most important

Because:

Missing a fraud case (False Negative) is very costly

Better to catch fraud even if some false alarms occur

3) How to handle class imbalance

Use resampling techniques:

Oversampling (SMOTE)

Undersampling

Use class weights

Adjust decision threshold

Use anomaly-focused evaluation metrics

4) High accuracy but low recall — what does it mean?

Model predicts normal cases well

But misses most fraud cases

Practically:

Bank losses increase

Fraudsters go undetected

Model is unsafe for real-world use

5) Complete ML Pipeline (End-to-End)

Data Preprocessing:
Handle missing values, remove noise

Feature Scaling:
Standardization / Normalization

Resampling Strategy:
SMOTE or class weighting

Model Training:
Train cost-sensitive model

Model Evaluation:
Recall, Precision, F1-score, ROC-AUC
